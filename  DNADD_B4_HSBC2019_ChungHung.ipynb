```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

#Task 2: In the adult data, the task is to predict the value of Income Group variable
#using other features.

# 1. Determine the type of supervised machine learning algorithm that should
# be used for task1 and task2 respectively [1M]

#Task 2
# Decision Tree, Random Forest

# 2. Choose any one of the above datasets to work with and import only that
# dataset into Python as a pandas data frame. [1M]. 
adult_data = pd.read_csv("adult.csv")
adult_data.head()
```

```python
# 3a . Determine the number of missing values in each column. The missing
# values are indicated using the character ‘?’ [1M].
# Note: You are not required to take any further action. Ensure that
# you retain the missing values as ‘?’ only. 

adult_data.replace('?',np.NaN)
adult_data.isnull().values.any()
```

```python
# 3b. Determine the data type of each column in the data and list out all
# columns that you think are categorical [1M]
features = adult_data.columns.drop("income_group")
category_features = ["workclass", "education", "marital_status", "occupation", 
                                        "relationship", "race", "sex", "native_count"]
category_target = ["income_group"]

# 3c. One hot encode the categorical columns such that k-1 dummy
# variables are created for k categories [1M]
    
adult_data_encoded = pd.get_dummies(adult_data[features])
adult_data_encoded["income_group"]= pd.get_dummies(adult_data[category_target])["income_group_ >50K"]

adult_data_encoded.head()
```

```python
# 3d. Scale all numerical columns such that after scaling the minimum
# value in the column is 0 and the maximum is 1. [1M]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features_to_scale = ["age","fnlwgt","education_num","capital_gain","capital_loss","hours_per_week"]
scaled_values = scaler.fit_transform(adult_data_encoded[features_to_scale])
adult_data_encoded["age"] = scaled_values[:,0]
adult_data_encoded["fnlwgt"] = scaled_values[:,1]
adult_data_encoded["education_num"] = scaled_values[:,2]
adult_data_encoded["capital_gain"] = scaled_values[:,3]
adult_data_encoded["capital_loss"] = scaled_values[:,4]
adult_data_encoded["hours_per_week"] = scaled_values[:,5]

adult_data_encoded.head()
```

```python
# 3ei. Comment on whether the data is adequately balanced [1M]
#ii. For the adult data check only the ‘income_group’ column 

#not balance enough, the income group < 50 K only got 7800 record

adult_data_encoded["income_group"].value_counts()
```

```python
# 3f. List out any 2 algorithms (or methods in sklearn) that can be used for
#making predictions on the chosen dataset. [1M]

# decision tree, random forest, logistic regression

# 3g. Split the data into train and test such that 20% of the total data is
#considered as test data [1M]

from sklearn.model_selection import train_test_split
adult_train, adult_test = train_test_split(adult_data_encoded, test_size=0.2)
adult_train.shape, adult_test.shape
```

```python
# 3h. Build a predictive machine learning model on the train data to
# predict the outcome (rings for abalone data, income_group for adult
# data) using an appropriate algorithm [2M]

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

model.fit(adult_train[adult_data_encoded.columns.drop("income_group")], adult_train["income_group"])
```

```python
# 3i. Evaluate the model’s performance on the train data using an
#appropriate metric. Note: choose among RMSE or Accuracy. [2M]

#Evaluate the model’s performance on the test data using an
#appropriate metric. Note: choose among RMSE or Accuracy. [2M]

train_accuracy = model.score(adult_train[adult_data_encoded.columns.drop("income_group")], adult_train["income_group"] )
test_accuracy = model.score(adult_test[adult_data_encoded.columns.drop("income_group")], adult_test["income_group"] )
print(train_accuracy, test_accuracy)
```

```python
#Comment on the usefulness of the model based on its performance
#in train and test data [2M]

# The model is useful because 
# the training accuracy and testing accuracy got 99.9% and 82.3% accuracy.
# Which is quite high
```

```python
#Finally build a model using an ensemble learner (either
#RandomForest or GradientBoosting) for your chosen data. Display
#the model’s score (RMSE or Accuracy) on both train and test data.
#[3M]

from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()

model1.fit(adult_train[adult_data_encoded.columns.drop("income_group")], adult_train["income_group"])

train_accuracy1 = model1.score(adult_train[adult_data_encoded.columns.drop("income_group")], adult_train["income_group"] )
test_accuracy1 = model1.score(adult_test[adult_data_encoded.columns.drop("income_group")], adult_test["income_group"] )
print(train_accuracy1, test_accuracy1)
```

```python
# The model is useful because 
# the training accuracy and testing accuracy got 98.6% and 84.8% accuracy.
# Which is quite high
```
