```python
#3. Study the target column distribution and state your insights.

#select features
features = bank_data.columns.drop("Personal Loan")
category_features =["ZIP Code","Education","Securities Account", "CD Account", "Online", "CreditCard"]
target = ["Personal Loan"]


#encoding the categorical variable
bank_data_encoded = pd.get_dummies(bank_data[features])
bank_data_encoded["Personal Loan"] = bank_data["Personal Loan"]

# from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# cat_columns_idx = [bank_data.columns.get_loc(col) 
#                    for col in category_features]
# ohe = OneHotEncoder(categorical_features=cat_columns_idx, 
#                     sparse=False, handle_unknown="ignore")
# bank_data_encoded = ohe.fit_transform(bank_data)

bank_data_encoded.head()
```

```python
#Normalize the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features_to_scale = ["Age", "Experience","Income","Family","CCAvg","Mortgage"]
scaled_values = scaler.fit_transform(bank_data_encoded[features_to_scale])
bank_data_encoded["Age"] = scaled_values[:,0]
bank_data_encoded["Experience"] = scaled_values[:,1]
bank_data_encoded["Income"] = scaled_values[:,2]
bank_data_encoded["Family"] = scaled_values[:,3]
bank_data_encoded["CCAvg"] = scaled_values[:,4]
bank_data_encoded["Mortgage"] = scaled_values[:,5]

#4. Split the data into training and test set in the ratio of 70:30 respectively
from sklearn.model_selection import train_test_split
bank_train, bank_test = train_test_split(bank_data_encoded, test_size=0.3)
bank_train.shape, bank_test.shape
```

```python
bank_data_encoded.head()
```

```python
#5. Use different classification models (Logistic, K-NN) to predict the likelihood of a liability
#customer buying personal loans (10 points)

#build the model using Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(bank_train[features],bank_train[target])

```

```python
#build the model using KNN
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier(n_neighbors=2,metric="euclidean")
model1.fit(bank_train[features],bank_train[target])
```

```python
#build the model using random forest
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
model2.fit(bank_train[features], bank_train[target])
```

```python
#6. Print the confusion matrix for all the above models (10 points)

#Confusion Matrix
from sklearn.metrics import confusion_matrix

train_predictions_LR = model.predict(bank_train[features])
test_predictions_LR = model.predict(bank_test[features])
train_conf_matrix_LR = confusion_matrix(bank_train[target], train_predictions_LR)
test_conf_matrix_LR = confusion_matrix(bank_test[target],test_predictions_LR)

#evaluate model performance on train and test set
train_accuracy = model.score(bank_train[features],bank_train[target])
test_accuracy = model.score(bank_test[features],bank_test[target])
print(train_accuracy, test_accuracy)

train_predictions_KNN = model1.predict(bank_train[features])
test_predictions_KNN = model1.predict(bank_test[features])
train_conf_matrix_KNN = confusion_matrix(bank_train[target], train_predictions_KNN)
test_conf_matrix_KNN = confusion_matrix(bank_test[target],test_predictions_KNN)


#evaluate model performance on train and test set
train_accuracy = model1.score(bank_train[features],bank_train[target])
test_accuracy = model1.score(bank_test[features],bank_test[target])
print(train_accuracy, test_accuracy)


train_predictions_RF = model2.predict(bank_train[features])
test_predictions_RF = model2.predict(bank_test[features])
train_conf_matrix_RF = confusion_matrix(bank_train[target], train_predictions_RF)
test_conf_matrix_RF = confusion_matrix(bank_test[target],test_predictions_RF)


#evaluate model performance on train and test set
train_accuracy = model2.score(bank_train[features],bank_train[target])
test_accuracy = model2.score(bank_test[features],bank_test[target])
print(train_accuracy, test_accuracy)
```
